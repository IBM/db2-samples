# ibm_db.num_fields()

## Purpose:

Determine the number of columns that exist in a result set.

## Syntax:

`int ibm_db.num_fields( IBM_DBStatement `*`statement`*` )`

## Parameters:

* __*statement*__&nbsp; &nbsp; : A valid `IBM_DBStatement` object that contains a result set.

## Return values:

* If __successful__, an integer representing the number of columns (fields) that exist in the result associated with the `IBM_DBStatement` object specified.
* If __unsuccessful__ or if the `IBM_DBStatement` object specified is not valid, the value `False`.

## Description:

The __ibm_db.num_fields()__ API is used to determine the number of columns that exist in a result set. Having this information can be useful when retrieving and processing result sets that were produced by dynamically generated queries, or were returned by stored procedures.
